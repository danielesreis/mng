```python
import cv2
import matplotlib.pyplot as plt
import os

plt.rcParams['figure.figsize'] = [20, 15]
```

```python
bgr = cv2.imread(os.getcwd() + '\\..\\segmented.jpg',0)
gray = cv2.cvtColor(bgr, cv2.COLOR_BGR2GRAY)
hist = cv2.calcHist([gray],[0],None,[256],[0,256])
plt.plot(hist)
```

```python
ret, mask = cv2.threshold(gray, 100, 255, cv2.THRESH_BINARY)
inv_mask = cv2.bitwise_not(mask)
thr_img = cv2.bitwise_and(gray, gray, mask = inv_mask)
thr_img2 = cv2.bitwise_and(bgr, bgr, mask = inv_mask)
thr_img2[thr_img2 == 0] = 255
```

```python
plt.imshow(thr_img2)
```

```python
cv2.imwrite(os.getcwd() + '//noshadow.jpg', thr_img2)
```
