```python
import pandas as pd
import re
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
import cv2
import numpy as np
import os
import math
```

```python
train_data = pd.read_csv('data/train.csv')
test_data = pd.read_csv('data/test.csv')
```

```python
train_data.info()
```

```python
test_data.info()
```

```python
# dropping irrelevant info
train_data.drop(columns=['PassengerId', 'Ticket', 'Cabin'], inplace=True)
test_data.drop(columns=['Ticket', 'Cabin'], inplace=True)
```

```python
# Way more women survived
train_data.groupby('Survived')['Sex'].value_counts(normalize=True)
```

```python
train_data.groupby('Survived')['Pclass'].value_counts()
```

```python
# new column for family members
train_data['family_num'] = train_data['Parch'] + train_data['SibSp']
test_data['family_num'] = test_data['Parch'] + test_data['SibSp']
```

```python
# if you're alone or is accompanied with lots of family members, you'll probably die
train_data.groupby('family_num')['Survived'].value_counts(normalize=True)
```

```python
# categorize family_num
interval = (-1, 0, 3, 11)

cats = ['is_alone', 'small_family', 'big_family']

test_data['family_cat'] = pd.cut(test_data.family_num, interval, labels=cats)
train_data['family_cat'] = pd.cut(train_data.family_num, interval, labels=cats)
```

```python
train_data.groupby('family_cat')['Survived'].value_counts(normalize=True)
```

```python
pd.crosstab(train_data.family_cat, train_data.Survived)
```

```python
train_data.drop(columns=['Parch', 'SibSp', 'family_num'], inplace=True)
test_data.drop(columns=['Parch', 'SibSp', 'family_num'], inplace=True)
```

```python
def get_title(name):
    title_search = re.search(' ([A-Za-z]+)\. ', name)
    if title_search:
        return title_search.group(1)
    return ""
```

```python
train_data['honorifics'] = train_data['Name'].apply(get_title)
test_data['honorifics'] = test_data['Name'].apply(get_title)
```

```python
train_data['honorifics'].value_counts()
```

```python
test_data['honorifics'].value_counts()
```

```python
train_data['honorifics'] = train_data['honorifics'].replace(['Mlle', 'Ms'], 'Miss')
train_data['honorifics'] = train_data['honorifics'].replace('Mme', 'Mrs')
train_data['honorifics'] = train_data['honorifics'].replace(['Dr', 'Rev', 'Major', 'Col', 'Don', 'Countess', \
                                                             'Jonkheer','Capt', 'Lady', 'Sir'], 'other')
```

```python
test_data['honorifics'] = test_data['honorifics'].replace('Ms', 'Miss')
test_data['honorifics'] = test_data['honorifics'].replace(['Dr', 'Rev', 'Col', 'Dona'], 'other')
```

```python
train_data.drop(columns=['Name'], inplace=True)
test_data.drop(columns=['Name'], inplace=True)
```

```python
pd.crosstab(train_data.honorifics, train_data.Survived)
```

```python
train_data.isnull().any()
```

```python
test_data.isnull().any()
```

```python
train_data['Embarked'].value_counts()
```

```python
train_data.loc[train_data.Embarked.isnull(), 'Embarked'] = 'S'
```

```python
train_data.loc[train_data.Age.isnull(), 'Age'] = train_data.groupby(['Sex','Pclass','honorifics']).Age.transform('median')
test_data.loc[test_data.Age.isnull(), 'Age'] = test_data.groupby(['Sex','Pclass','honorifics']).Age.transform('median')
```

```python
# categorize Age
interval = (0, 1, 5, 12, 17, 26, 50, 65, 85)

cats = ['new_born', 'baby', 'kid', 'teen', 'young', 'adult', 'middle_age', 'elderly']

test_data["Age_cat"] = pd.cut(test_data.Age, interval, labels=cats)
train_data['Age_cat'] = pd.cut(train_data.Age, interval, labels=cats)
```

```python
print(pd.crosstab(train_data.Age_cat, train_data.Survived))

#Seting the figure size
plt.figure(figsize=(12,5))

#Plotting the result
sns.countplot("Age_cat",data=train_data,hue="Survived", palette="hls")
plt.xlabel("Categories names", fontsize=18)
plt.xlabel("Count", fontsize=18)
plt.title("Age Distribution ", fontsize=20)
plt.show()
```

```python
train_data.drop(columns=['Age'], inplace=True)
test_data.drop(columns=['Age'], inplace=True)
```

```python
test_data.loc[test_data.Fare.isnull(), 'Fare'] = test_data.groupby(['Sex','Pclass','honorifics']).Fare.transform('median')
```

```python
# the richest had higher chances to survive
train_data.groupby('Survived')['Fare'].mean()
```

```python
# categorize Fare
quant = (-1, 0, 8, 15, 31, 1000)
label_quants = ['NoInf', 'quart_1', 'quart_2', 'quart_3', 'quart_4']

train_data["Fare_cat"] = pd.cut(train_data.Fare, quant, labels=label_quants)
test_data["Fare_cat"] = pd.cut(test_data.Fare, quant, labels=label_quants)

train_data['Fare_cat'] = train_data['Fare_cat']
test_data['Fare_cat'] = test_data['Fare_cat']
```

```python
pd.crosstab(train_data.Fare_cat, train_data.Survived)
```

```python
sns.countplot(x="Fare_cat", hue="Survived", data=train_data, palette="hls")
plt.title("Count of survived x Fare expending",fontsize=20)
plt.xlabel("Fare Cat",fontsize=15)
plt.ylabel("Count",fontsize=15)

plt.show()
```

```python
train_data.drop(columns='Fare', inplace=True)
test_data.drop(columns='Fare', inplace=True)
```

```python
train_data.head()
```

```python
train_data = pd.get_dummies(train_data, columns=['Pclass','Sex','Embarked','family_cat','Age_cat','Fare_cat','honorifics'],\
                          prefix=['Pcl','Sex','Emb','Fam', 'Age','Fare','Prf'])
```

```python
test_data = pd.get_dummies(test_data, columns=['Pclass','Sex','Embarked','family_cat', 'Age_cat','Fare_cat','honorifics'],\
                         prefix=['Pcl','Sex','Emb','Fam', 'Age','Fare','Prf'])
```

```python
train_data.head()
```

```python
plt.figure(figsize=(20,12))
plt.title('Correlation of Features for Train Set')
sns.heatmap(train_data.astype(float).corr(),vmax=1.0,  annot=True)
plt.show()
```

```python
# dropping correlated features
train_data.drop(columns=['Pcl_1','Sex_male','Emb_S','Fam_is_alone','Age_adult','Prf_Mr'], inplace=True)
test_data.drop(columns=['Pcl_1','Sex_male','Emb_S','Fam_is_alone','Age_adult','Prf_Mr'], inplace=True)
```

```python
params = {'max_depth': [10, 11, 12, 13, 14, 15],\
          'min_samples_split': [5, 6, 7, 8, 9, 10],\
          'min_samples_leaf': [2, 3, 4, 5],\
          'n_estimators': [10, 25, 50, 75, 100]}
```

```python
clf = RandomForestClassifier()
```

```python
forest_cv = GridSearchCV(estimator=clf,param_grid=params, cv=5)
```

```python
forest_cv.fit(train_data.drop(columns=['Survived']), train_data['Survived'])
```

```python
forest_cv.best_score_
```

```python
forest_cv.best_estimator_
```

```python
y_pred = forest_cv.predict(test_data.drop(columns=['PassengerId']))
```
