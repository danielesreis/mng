```python
import cv2
import matplotlib.pyplot as plt
import os
```

```python
plt.rcParams['figure.figsize'] = [20, 15]
```

```python
bgr = cv2.imread('teste.jpg')

imgray = cv2.cvtColor(bgr,cv2.COLOR_RGB2GRAY)
ret,thresh = cv2.threshold(imgray,127,255,0)
image, contours, hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
```

```python
width, height, _ = bgr.shape
min_x, min_y = width, height
max_x = max_y = 0
```

```python
for contour in contours:
    (x,y,w,h) = cv2.boundingRect(contour)
    #print("x: " + str(x) + " y: " + str(y) + " w: " + str(w) + " h: " + str(h))
    if x != 0 and w != width:    
        min_x, max_x = min(x, min_x), max(x+w, max_x) 
        
    if y != 0 and h != height:
        min_y, max_y = min(y, min_y), max(y+h, max_y)
```

```python
print('min_x:' + str(min_x) + " max_x: " + str(max_x) + ' min_y: ' + str(min_y) + ' max_y: ' + str(max_y))
```

```python
bgr = cv2.rectangle(bgr, (min_x, min_y), (max_x, max_y), (0,255,0),2)
```

```python
plt.imshow(bgr)
```
