```python
import cv2
import matplotlib.pyplot as plt
import os

plt.rcParams['figure.figsize'] = [20, 15]
```

```python
bgr = cv2.imread('noshadow.jpg')

imgray = cv2.cvtColor(bgr,cv2.COLOR_RGB2GRAY)
ret,thresh = cv2.threshold(imgray,127,255,0)
image, contours, hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
```

```python
# checking the appropriate contour
i=1
for contour in contours:
    new_bgr = bgr.copy()
    (x, y, w, h) = cv2.boundingRect(contour)
    cv2.rectangle(new_bgr, (x,y), (x+w,y+h), (0,255,0),2)
    # cv2.imwrite(os.getcwd() + '\\teste\\' + str(i) + '.jpg', new_bgr)
    i = i + 1
```

```python
contour = contours[34]
(x,y,w,h) = cv2.boundingRect(contour)
# print("x: " + str(x) + " y: " + str(y) + " w: " + str(w) + " h: " + str(h))
plt.imshow(bgr[y:y+h,x:x+w])
```
